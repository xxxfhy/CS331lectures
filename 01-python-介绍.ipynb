### `if`-`else` è¯­å¥

### `while`å¾ªç¯

```python
f0 = 0
f1 = 1
while f0 < 100:
    print(f0)
    f0, f1 = f1, f0+f1
```

```python
to_find = 55
found = False

f0 = 0
f1 = 1
while f0 <= to_find:
    if to_find == f0:
        print(f'{to_find} is a Fibonacci number!')
        found = True
        break
    f0, f1 = f1, f0+f1

if not found:
    print(f'{to_find} is not a Fibonacci number!')
```

ä¸ºäº†ç®€åŒ–ä¸Šé¢çš„å®ç°æ¨¡å¼ï¼Œæˆ‘ä»¬å¯ä»¥åœ¨ `while` å¾ªç¯æœ¬èº«é™„åŠ ä¸€ä¸ª `else` å­å¥ã€‚åªæœ‰å½“ `while` å¾ªç¯æ²¡æœ‰æå‰ç»ˆæ­¢æ—¶(ä¾‹å¦‚ï¼Œé€šè¿‡ `break` æˆ– `return`) ï¼Œæ‰ä¼šè®¡ç®—è¿™ä¸ª `else` å­å¥

```python
to_find = 55

f0 = 0
f1 = 1
while f0 <= to_find:
    if to_find == f0:
        print(f'{to_find} is a Fibonacci number!')
        break
    f0, f1 = f1, f0+f1
else:
    print(f'{to_find} is not a Fibonacci number!')
```

### Dicts

[dictionary](https://docs.python.org/3/library/stdtypes.html#mapping-types-dict) æ˜¯ä¸€ä¸ªåŒ…å«ä¸€ç»„å”¯ä¸€é”®å€¼æ˜ å°„çš„æ•°æ®ç»“æ„ã€‚

key &rarr; value

```python
d = {
    'Superman':  'Clark Kent',
    'Batman':    'Bruce Wayne',
    'Spiderman': 'Peter Parker',
    'Ironman':   'Tony Stark'
}
```

```python
d['Ironman']
```

```python
d['Ironman'] = 'James Rhodes'
d
```

```python
del d['Ironman']
d
```

```python
for k in d:
    print(f'{k} => {d[k]}')
```

```python
for k in d.keys():
    print(f'{k} => {d[k]}')
```

```python
for v in d.values():
    print(v)
```

```python
for k, v in d.items():
    print(f'{k} => {v}')
```
